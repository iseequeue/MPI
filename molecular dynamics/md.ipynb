```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=np.array([983, 546, 388, 299])
p=np.array([1, 2, 3, 4])
s=[]
e=[]
for j in range(len(t)):
  s.append(t[0]/t[j])
  e.append(t[0]/(t[j]*(j+1)))
```

```python
fig = plt.figure(figsize=(20, 6))

ax = plt.subplot(1, 3, 1)
plt.plot(p, t, c='blue', marker = 'o')
plt.ylabel("t, seconds", fontsize=18, fontweight="bold")
plt.xlabel('p', fontsize=18, fontweight="bold")
plt.title("Время работы", fontsize=18, fontweight="bold")

ax = plt.subplot(1, 3, 2)
plt.plot(p, s, label='Ускорение', c='red', marker = 'o')
plt.ylabel("s", fontsize=18, fontweight="bold")
plt.xlabel('p', fontsize=18, fontweight="bold")
plt.title("Ускорение", fontsize=18, fontweight="bold")

ax = plt.subplot(1, 3, 3)
plt.plot(p, e, label='Эффективность', c='green', marker = 'o')
plt.ylabel("e", fontsize=18, fontweight="bold")
plt.xlabel('p', fontsize=18, fontweight="bold")
plt.title("Эффективность", fontsize=18, fontweight="bold")
```

```python
f = open('/content/energy.txt')
energy=[]
for line in f:
  energy.append(float(line))

x = [i for i in range(len(energy))]
import matplotlib.pyplot as plt
import numpy as np
q = np.array(energy)/500

f2 = open('/content/kinetic.txt')
kinetic=[]
for line in f2:
  kinetic.append(float(line))

x = [i for i in range(len(energy))]
import matplotlib.pyplot as plt
import numpy as np
qq=np.array(kinetic)/500

fig = plt.figure(figsize=(20, 10))

plt.tick_params(axis='both', which='major', labelsize=14)

plt.title('Full energy', fontsize=18, fontweight="bold")  
plt.xlabel('N', fontsize=18, fontweight="bold")
plt.ylabel('E/N', fontsize=18, fontweight="bold")
plt.plot(x, q,  c="r")
#plt.plot(x, qq,  c="b")
```

```python
fig = plt.figure(figsize=(20, 10))

plt.tick_params(axis='both', which='major', labelsize=14)

plt.title('Kinetic energy', fontsize=18, fontweight="bold")  
plt.xlabel('N', fontsize=18, fontweight="bold")
plt.ylabel('E/N', fontsize=18, fontweight="bold")
#plt.plot(x, q,  c="r")
plt.plot(x, qq,  c="b")
```
