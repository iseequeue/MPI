## 1. π 
Построим зависимость времени работы Tp, ускорения Sp и эффективности Ep в зависимости от числа процессоров p и числа разбиений n.

```python
n=np.array([100000000, 200000000, 300000000, 400000000, 500000000])
t=[]
s=[]
e=[]
t.append(np.array([3.05413,6.15294,9.15273, 12.5988, 15.6785]))
t.append(np.array([1.65891,3.35179,5.05424,6.85823,8.2957]))
t.append(np.array([1.25566,2.50883,3.72195,5.09083,6.38874]))
t.append(np.array([1.05556,2.07688,3.18005,4.22802,5.20945]))


for j in range(len(t)):
  s.append(t[0]/t[j])
  e.append(t[0]/(t[j]*(j+1)))
```

```python
fig = plt.figure(figsize=(20, 10))

# Tp

#ax = plt.subplot(1, 3, 1)
plt.tick_params(axis='both', which='major', labelsize=14)


plt.plot(n, t[0], label='Время работы', c='b', marker = 'o')
plt.plot(n, t[1], label='Время работы', c='r', marker = 'o')
plt.plot(n, t[2], label='Время работы', c='y', marker = 'o')
plt.plot(n, t[3], label='Время работы', c='g', marker = 'o')

plt.ylabel("Tp", fontsize=18, fontweight="bold")
plt.xlabel('n', fontsize=18, fontweight="bold")
plt.title("Время работы", fontsize=18, fontweight="bold")
plt.grid(alpha=0.2)
plt.legend(['p=1', 'p=2', 'p=3', 'p=4'])
plt.ylabel("Tp", fontsize=18, fontweight="bold")
plt.xlabel('n', fontsize=18, fontweight="bold")


plt.show()
```

```python
fig = plt.figure(figsize=(20, 10))

# Tp
def Tp(n,p):
  return 6*n/p+6+log2(p)

#ax = plt.subplot(1, 3, 1)
plt.tick_params(axis='both', which='major', labelsize=14)

y=[t[0][3], t[1][3], t[2][3], t[3][3] ]
x=np.array([1,2,3,4])

y=[t[0][0], t[1][0], t[2][0], t[3][0] ]
plt.plot(x, y, label='Ускорение', c='b', marker = 'o')
y=[t[0][1], t[1][1], t[2][1], t[3][1] ]
plt.plot(x, y, label='Ускорение', c='r', marker = 'o')
y=[t[0][2], t[1][2], t[2][2], t[3][2] ]
plt.plot(x, y, label='Ускорение', c='y', marker = 'o')
y=[t[0][3], t[1][3], t[2][3], t[3][3] ]
plt.plot(x, y, label='Ускорение', c='g', marker = 'o')

y=[t[0][4], t[1][4], t[2][4], t[3][4] ]
plt.plot(x, y, label='Ускорение', c='black', marker = 'o')

n=t[0][3]/6
plt.plot(x, Tp(n,x), label='Ускорение', c='purple', marker = 'o')

plt.legend(['p=1e8', 'p=2e8', 'p=3e8', 'p=4e8', 'p=5e8'])
plt.title("Время работы", fontsize=18, fontweight="bold")
plt.ylabel("Tp", fontsize=18, fontweight="bold")
plt.xlabel('p', fontsize=18, fontweight="bold")
```
