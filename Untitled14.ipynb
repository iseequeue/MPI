## 2. Проведем расчет метода Монте Карло с потенциалом Эвальда

В качестве начальной конфигурации бралась случайная конфигурация из 30 частиц, число шагов алгоритма Метрополиса 10 000

```python
def pogr(q, start, N):

  e = []
  for i in range(N):

    k1 = int(i*(len(q)-start)/N) + start
    k2 = int((i+1)*(len(q) - start)/N) + start
    e.append(q[k1:k2-1:].mean())

  e = np.array(e)
  return e.std()
```

```python
def v22(x, n):
  return 0.5*np.sin(2*pi*n*x)/(pi*n*x)

def f(n):
  return np.exp(-pi*n**2) - pi * n * math.erfc(n* pi **0.5)

def v21(x, n):
    return (f(abs(n-x)) - f(abs(n+x)))/(pi * n * x)
```

```python
def F1(x):
  return 1/x - 1.76011884*(4*pi/3)**(1/3) + 2* pi *x**2 /3

def F2(x):
  y=0.0
  N=5 #max n
  for i in range(N):
    for j in range (N):
      for k in range(N):
        n = (i*i+j*j+k*k)**0.5
        if n<=1e-6:
          continue
        y+=0.5*(v21(x, n)+np.exp(-pi*n*n)/(pi*n*n)*v22(x, n))
  return math.erfc(pi**0.5 * x)/x - 1 + y
```

```python
plt.figure(figsize=(13, 9))
x=np.linspace(-2, 2, 100)
plt.plot(x, F1(x),  c="b")
y=[F2(i) for i in x]
plt.scatter(x,y,  c="r")
plt.legend(['1/x - c0 + 2pix^2/3', 'длинный ряд'])
plt.grid()
```
