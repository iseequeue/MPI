```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=np.array([0.150402, 0.114571, 0.0911457, 0.0717284])
p=np.array([1, 2, 3, 4])
s=[]
e=[]
for j in range(len(t)):
  s.append(t[0]/t[j])
  e.append(t[0]/(t[j]*(j+1)))
```

```python
fig = plt.figure(figsize=(20, 6))

ax = plt.subplot(1, 3, 1)
plt.plot(p, t, c='blue', marker = 'o')
plt.ylabel("t, seconds", fontsize=18, fontweight="bold")
plt.xlabel('p', fontsize=18, fontweight="bold")
plt.title("Время работы", fontsize=18, fontweight="bold")

ax = plt.subplot(1, 3, 2)
plt.plot(p, s, label='Ускорение', c='red', marker = 'o')
plt.ylabel("s", fontsize=18, fontweight="bold")
plt.xlabel('p', fontsize=18, fontweight="bold")
plt.title("Ускорение", fontsize=18, fontweight="bold")

ax = plt.subplot(1, 3, 3)
plt.plot(p, e, label='Эффективность', c='green', marker = 'o')
plt.ylabel("e", fontsize=18, fontweight="bold")
plt.xlabel('p', fontsize=18, fontweight="bold")
plt.title("Эффективность", fontsize=18, fontweight="bold")
```
